```python
county_large_df = sorted_school_df.groupby("County Name")["Total"].sum().sort_values(ascending=False).head(10) 
county_large_df
```

```python
plt.barh(county_large_df.index, county_large_df.values)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.pie(
    gender_school_df,
    labels=gender_school_df.index,
    startangle=125,
    colors=[ "#A157DD", "#64DEA1"],
    autopct="%1.1f%%"
)

plt.title("Male/Female Absenteeism Rate")

plt.legend(
    labels = gender_school_df.index,
    loc="upper left",
    # title = "Platform"
)

plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.scatter(school_df["Female"], school_df["Male"])

plt.title("Male vs Female Counts")
plt.xlabel("Female")
plt.ylabel("Male")

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
